```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
from sklearn.datasets import fetch_california_housing
boston_data=fetch_california_housing()
```

```python
print(boston_data.keys())
```

```python
boston=pd.DataFrame(boston_data.data,columns=boston_data.feature_names)
```

```python
boston.head()
```

```python
boston.isnull().sum()
```

```python
boston.info()
```

```python
boston.drop(columns=['Latitude','Longitude'])
```

```python
boston['MHV']=boston_data.target
```

```python
boston
```

```python
plt.hist(boston,bins=4)
plt.show()
```

```python
sns.distplot(boston['MHV'],bins=20)
plt.show()
```

```python
sns.distplot(boston['MHV'],bins=20)

corln=boston.corr().round(2)
sns.heatmap(data=corln,annot=True)
```

```python
plt.figure(figsize=(20,5))
feature=['AveBedrms','Population']
targ=boston['MHV']
```

```python
for i,col in enumerate(feature):
    plt.subplot(1,len(feature),i+1)
    x=boston[col]
    y=targ
    plt.scatter(x,y,marker='o')
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('MHV')
```

```python
from sklearn.model_selection import train_test_split
```

```python
Population=boston['Population'].values.reshape(-1,1)
MHV=boston['MHV'].values.reshape(-1,1)
plt.scatter(MHV,Population)
```

```python
x_train, x_test, y_train, y_test=train_test_split(MHV,Population,test_size=0.3,random_state=100)
x_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
reg.intercept_
```

```python
pred_1=reg.predict(x_test)
```

```python
pred_1
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.metrics import r2_score
```

```python
mae=mean_absolute_error(y_test,pred_1)
mse=mean_squared_error(y_test,pred_1)
score=r2_score(y_test,pred_1)
```
